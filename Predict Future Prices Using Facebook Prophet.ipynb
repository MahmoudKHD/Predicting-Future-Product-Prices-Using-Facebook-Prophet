# TASK 4: PREPARE THE DATA BEFORE APPLYING FACEBOOK PROPHET TOOL 

```python
avocado.head()
```

```python
avocado_prophet=avocado[['Date', 'AveragePrice']]
```

```python
avocado_prophet.head()
```

```python
avocado_prophet= avocado_prophet.rename(columns={'Date':'ds', 'AveragePrice':'y'})
```

```python
avocado_prophet.head()
```

# TASK 5: UNDERSTAND INTUITION BEHIND FACEBOOK PROPHET

# TASK 6: DEVELOP MODEL AND MAKE PREDICTIONS - PART A

```python
m= Prophet()
m.fit(avocado_prophet)
```

```python
# Forcasting into the future
future= m.make_future_dataframe(periods=365)
forecast=m.predict(future
                  )
```

```python
forecast
```

```python
figure=m.plot(forecast, xlabel='Date', ylabel='Price')
```

```python
figure2=m.plot_components(forecast)
```

# TASK 7: DEVELOP MODEL AND MAKE PREDICTIONS (REGION SPECIFIC) - PART B

```python
# dataframes creation for both training and testing datasets 
avocado = pd.read_csv('avocado.csv')
```

```python
# Select specific region
avocado_sample=avocado[avocado['region']=='Chicago']
```

```python
avocado_sample=avocado_sample.sort_values('Date')
```

```python
plt.plot(avocado_sample['Date'], avocado_sample['AveragePrice'])
```

```python
avocado_sample=avocado_sample.rename(columns={'Date':'ds', 'AveragePrice':'y'})
```

```python
m = Prophet()
m.fit(avocado_sample)
# Forcasting into the future
future = m.make_future_dataframe(periods=365)
forecast = m.predict(future)
```

```python
figure = m.plot(forecast, xlabel='Date', ylabel='Price')
```

```python
figure3 = m.plot_components(forecast)
```
